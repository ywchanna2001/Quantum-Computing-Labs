```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
```

```python
qc = QuantumCircuit(1)
qc.h(0)             # Hadamard gate
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
```

```python
qc = QuantumCircuit(5)

qc.h(0)
qc.x(1)
qc.y(4)
qc.barrier()      

qc.cx(0,1)
qc.cx(1,2)

qc.y(0)
qc.h(1)
qc.y(2)
qc.x(3)
qc.h(4)

qc.cx(2,3)
qc.cx(2,4)
qc.cx(3,4)

qc.h(0)
qc.x(4)

qc.cx(4,0)
qc.draw("mpl")
```

```python
qc = QuantumCircuit(5)
qc.h(0)
qc.x(1)
qc.y(4)
qc.barrier()

qc.cx(0,1)
qc.barrier()

qc.cx(1,2)
qc.barrier()

qc.draw("mpl")
```

2-Qubit quantum circuit

```python
qc = QuantumCircuit(2,2)    # 2 -> two qubits, 2 -> two classical bits
qc.h(0)
qc.cx(0,1)
qc.measure([0,1], [0,1])        # two classical bits for storing results of measurement
qc.draw("mpl")
```

```python
from qiskit_aer import Aer
```

```python
print(Aer.backends())
```

```python
sim_backend = Aer.get_backend("qasm_simulator")
```

```python
from qiskit import transpile
```

```python
compiled = transpile(qc, sim_backend)
```

```python
job = sim_backend.run(compiled)
```

```python
results = job.result()
```
